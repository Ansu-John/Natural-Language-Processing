### Frequency Distributions

```python
# get frequency of each vocabulary item in the text
fdist1 = FreqDist(text1)
print(fdist1)
```

```python
fdist1.most_common(5)
```

```python
# count of the number of times a given sample occurred
fdist1['the']
```

```python
# frequency of a given sample
fdist1.freq('the')
```

```python
# total number of samples
fdist1.N()
```

```python
#tabulate the frequency distribution
fdist1.tabulate()
```

```python
# graphical plot of the frequency distribution
fdist1.plot()
```

```python
# cumulative plot of the frequency distribution
fdist1.plot(cumulative=True)
```

```python
# find the words from the vocabulary of the text that are more than 15 characters long
V = set(text1)
long_words = [w for w in V if len(w) > 15]
sorted(long_words)
```

### Collocations and Bigrams

A collocation is a sequence of words that occur together unusually often. Thus red wine is a collocation, whereas the wine is not. A characteristic of collocations is that they are resistant to substitution with words that have similar senses; for example, maroon wine sounds definitely odd. To get a handle on collocations, we start off by extracting from a text a list of word pairs, also known as bigrams. 

```python
list(bigrams(['more', 'is', 'said', 'than', 'done']))
```

```python
text4.collocations()
```

### Find the distribution of word lengths in a text

```python
# creating a FreqDist out of a long list of numbers, where each number is the length of the corresponding word in the text
[len(w) for w in text1]
```

```python
# create a frequency distribution containing the given samples
fdist = FreqDist(len(w) for w in text1)
print(fdist)
```

```python
fdist
```

```python
# the n most common samples and their frequencies
fdist.most_common()
```

```python
# sample with the greatest count
fdist.max()
```
